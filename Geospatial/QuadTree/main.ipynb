## Build a QuadTree
```python
QuadTree(n, k, w, h)

Args:
            n (int): The number of points to be stored in the quadtree.
            k (int): The number of points per leaf node.
            w (int): The width of the quadtree.
            h (int): The height of the quadtree.
```
Note that 
* All nodes in QuadTree are generated randomly.
* Because the ratio of the graph represented below is not exactly 1:1 (actually is 3:2), the distance relations between pairs of points are not apparent, which indicates that a circle in the graph seems similar to an ellipse.

````python
from quadtree import QuadTree

tree = QuadTree(200, 3, 50, 50)

tree.build()

tree.visualize()
````
